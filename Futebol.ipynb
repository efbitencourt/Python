```python
#Separando a contagem de partidas em casa, por país
resultados_copa_novo['home_team'].value_counts()
```

```python
#Separando a contagem de partidas em casa, por país, outra forma de fazer
ranking_casa = resultados_copa_novo.groupby(['home_team']).size().sort_values(ascending = False)
```

```python
ranking_casa = pd.DataFrame(ranking_casa)
```

```python
ranking_casa = ranking_casa.reset_index()
```

```python
ranking_casa
```

```python
#Separando a contagem de partidas fora, por país
ranking_visita = resultados_copa_novo.groupby(['away_team']).size().sort_values(ascending = False)
```

```python
ranking_visita = ranking_visita.reset_index()
```

```python
#Deixando as duas com o mesmo nome de coluna, para conseguir fazer o merge depois
ranking_visita.rename(columns={'away_team':'home_team'})
```

```python
#Deixando as duas com o mesmo nome de coluna, para conseguir fazer o merge depois
ranking_visita = ranking_visita.rename(columns={'away_team':'home_team'})
ranking_visita
```

```python
#Unindo os dois dataframes (casa e fora)
ranking_partidas = ranking_casa.merge(ranking_visita, how='outer', on='home_team')
```

```python
ranking_partidas.sort_values(by='0_x', ascending=False)
```

```python
#Criando uma nova coluna com a soma das partidas em casa e fora
ranking_partidas['partidas'] = ranking_partidas['0_x'] + ranking_partidas['0_y']
```

```python
#Ordenando o ranking
ranking_partidas.sort_values(by='partidas', ascending=False)
```

```python
#Duas situações precisam ser resolvidas
# 1 - converter o número de partidas de string para inteiro
# 2 - Substituir os nulos por zero para poder ajustar a soma
```

```python
#Substituindo zeros por nulo
ranking_partidas = ranking_partidas.fillna(0)
ranking_partidas
```

```python
ranking_partidas = ranking_partidas.sort_values(by='partidas', ascending=False)
```

```python
ranking_partidas
```

```python
ranking_partidas['partidas'].describe()
```

```python
#Renomeando as colunas
ranking_partidas = ranking_partidas.rename(columns={'0_x':'partidas_mandante', '0_y':'partidas_visitante'})
ranking_partidas                                                    
```

```python
#convertendo número de partidas para inteiro
ranking_partidas['partidas_mandante'] = ranking_partidas['partidas_mandante'].astype(int)
ranking_partidas['partidas_visitante'] = ranking_partidas['partidas_visitante'].astype(int)
ranking_partidas['partidas'] = ranking_partidas['partidas'].astype(int)
```

```python
#reorganizando o index para colocar na ordem do ranking
ranking_partidas = ranking_partidas.reset_index()
ranking_partidas
```

```python
#removendo as colunas desnecessárias
ranking_partidas = ranking_partidas.drop(columns=['index','partidas_mandante','partidas_visitante'])
ranking_partidas
```

```python
#ajustando o nome da coluna do país
ranking_partidas.rename(columns={'home_team':'País'}, inplace=True)
```
