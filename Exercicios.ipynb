```python
bd_teste.drop(['Categoria'], axis=1, inplace=True)
```

```python
bd_teste['Categoria de idade'] = bd_teste['housing_median_age'].apply(lambda x: "Imóvel novo" if x <= 20 \
                                                                      else "Imóvel meia idade" if x <= 35 \
                                                                      else "Imóvel antigo")
```
